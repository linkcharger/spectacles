## pdf plumber
https://github.com/jsvine/pdfplumber

```python
import pdfplumber

with pdfplumber.open('data/1939/sample.pdf') as pdf:
    page = pdf.pages[0]

    left = page.crop((0, 0, 0.5 * page.width, 0.9 * page.height))
    right = page.crop((0.5 * page.width, 0, page.width, page.height))

    print(left.extract_text(
        x_tolerance=1, 
        y_tolerance=2,
        
        layout=False).replace('\n', ' '))
```

## pyMuPDF
https://pymupdf.readthedocs.io/en/stable/index.html

```python
import fitz # install using: pip install PyMuPDF

with fitz.open('data/1939/sample.pdf') as doc:
    print(doc[0].get_text(sort=True).replace('\n', ''))
```

```python
with fitz.open('data/1939/sample.pdf') as doc:
    page = doc[0]
    textPage = page.get_textpage_ocr(flags=3, language='eng', dpi=200, full=True)

    print(
        ' '.join(
            [block[4] for block in page.get_text('blocks', textpage=textPage, sort=True)]
        )
    )
```
